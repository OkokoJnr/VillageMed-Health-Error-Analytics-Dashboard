```python
import pandas as pd
```

```python
#import cleaned dataset
Nigeria_medical_error = pd.read_csv("../cleaned_data/cleaned_medical_error_records_Nigeria.csv")
Other_medical_error = pd.read_csv("../cleaned_data/cleaned_medical_error_records_Others.csv")
```

```python
Nigeria_medical_error.head()
```

```python
Other_medical_error.head()
```

```python
#Join dataframes
combined_medical_error = pd.concat([Nigeria_medical_error, Other_medical_error], ignore_index=True)
combined_medical_error.info()
```

```python
#distribution of medical error
count_error_type = combined_medical_error['error_type'].value_counts().sort_values(ascending=True)
#count_error_type.plot(kind='barh', title='Distribution of Medical Error Types', xlabel='Count of Error Type', ylabel='Error Types');
```

```python
#combined_medical_error[['error_type', 'country']].value_counts().sort_values(ascending=True)
combined_medical_error.groupby('country')['error_type'].value_counts().unstack().fillna(0)
```

```python
#Distribution by facilities
count_facility = combined_medical_error["hospital_facility_type"].value_counts().sort_values(ascending=True)
count_facility
```

```python
#Distribution by location
count_location = combined_medical_error["country"].value_counts().sort_values(ascending=True)
count_location
```

```python
#export combined dataset 
combined_medical_error.to_csv("../cleaned_data/combined_medical_error_records.csv", index=False)
```
